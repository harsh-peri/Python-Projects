```python
import matplotlib.pyplot as plt  
import matplotlib.image as mpimg  
import numpy as np

image_path = r"pic2.jpg" 

try:  
    img = mpimg.imread(image_path)  

    print("Image Dimensions:", img.shape)  

    print("Is Image Integer Type?:", np.issubdtype(img.dtype, np.integer))  

    plt.imshow(img)  
    plt.axis('off')  
    plt.show()  

except FileNotFoundError:  
    print(f"Error: The file '{image_path}' was not found. Please check the path.")  
except Exception as e:  
    print(f"An unexpected error occurred: {e}")  
```

```python
# 4. Convert to Grayscale
gray_img = np.dot(img[..., :3], [0.2989, 0.5870, 0.1140])
plt.figure(figsize=(6,6))
plt.imshow(gray_img, cmap='gray')
plt.axis('off')
plt.title("Grayscale Image")
plt.show()
```

```python
# 5. Lightening & Darkening the Image
lighter_img = np.clip(gray_img + 50, 0, 255)
darker_img = np.clip(gray_img - 50, 0, 255)

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(lighter_img, cmap='gray')
plt.axis('off')
plt.title("Lighter Image")

plt.subplot(1,2,2)
plt.imshow(darker_img, cmap='gray')
plt.axis('off')
plt.title("Darker Image")
plt.show()
```

```python
# 6. Adjust Contrast
contrast_img = np.clip((gray_img - 128) * 1.5 + 128, 0, 255)

plt.figure(figsize=(6,6))
plt.imshow(contrast_img, cmap='gray')
plt.axis('off')
plt.title("Contrast Adjusted Image")
plt.show()
```

```python
# 7. Crop the Image
h, w = gray_img.shape
cropped_img = gray_img[int(h*0.3):int(h*0.5), int(w*0.3):int(w*0.5)]

plt.figure(figsize=(5.5,5.5))
plt.imshow(cropped_img, cmap='gray')
plt.axis('off')
plt.title("Cropped Image")
plt.show()
```

```python
#8.apply Matrix Transformations
flipped_img = np.fliplr(gray_img)

rotated_img = np.rot90(gray_img)

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(flipped_img, cmap='gray')
plt.axis('off')
plt.title("Flipped Image")

plt.subplot(1,2,2)
plt.imshow(rotated_img, cmap='gray')
plt.axis('off')
plt.title("Rotated Image")
plt.show()
```
