```python
import customtkinter as ctk
import tkinter as tk
from tkinter import ttk
import tkcalendar as tkcal
import wikipedia as wp
import datetime as dt
import mysql.connector
from tkinter import messagebox
from PIL import Image

class Application:  
    def __init__(self, master):
        self.master=master
        self.master.geometry("{w}x{h}+0+0".format(w=self.master.winfo_screenwidth()-10, h=self.master.winfo_screenheight()-10))
        #self.master.geometry("500x500")
        self.master.iconbitmap("F:/source/get help son/mendiog-ico.ico")
        self.master.title("MENDI - The Mental Wellness App")
        self.master._set_appearance_mode("System")
        self.home=HomeFrame(self.master)
        self.home.pack(side='right', fill="both", expand=1)
    db=mysql.connector.connect(
    host="localhost",
    user="MendiAdmin",
    passwd="***************************",
    database="MendiDB")
    c=db.cursor()

class MenuFrame(ctk.CTkFrame):
    def __init__(self, master):
        killallframes()
        ctk.CTkFrame.__init__(self, master, fg_color="#000", corner_radius=0)
        self.master=master
        self.label=ctk.CTkLabel(self, text="Mendi", text_color="#b795ff", corner_radius=10, font=ctk.CTkFont("Brush Script MT", 30, weight='bold'), fg_color='transparent', height=37)
        self.label.pack(anchor='center')
        self.menubutton=ctk.CTkButton(self, fg_color='transparent', text="Menu", width=80, hover_color="#b795ff")
        self.menubutton.place(x=2, rely=0.10, in_=self)
        def createdrop(self):
            self.dropdown=DropFrame(self.master)
            self.dropdown.focus_set()
            self.dropdown.bind("<Enter>", lambda event: [self.dropdown.pack(side="left", fill="y")])
            self.dropdown.bind("<Leave>", lambda event: [self.dropdown.destroy(), createdrop(self)])
        createdrop(self)
        self.menubutton.bind("<Enter>", lambda event: [self.dropdown.pack(side="left", fill="y")])
        #self.button.bind("<Leave>", lambda event:[self.menu.destroy(), createdrop(self)])

class HomeFrame(ctk.CTkFrame):
    def __init__(self, master):
        killallframes()
        self.master=master
        self.menu=MenuFrame(self.master)
        self.menu.pack(side='top', fill='x')
        ctk.CTkFrame.__init__(self, master, fg_color='#3d113d', corner_radius=0)
        self.rightframe=ctk.CTkFrame(self, fg_color="#7860ac", height=700, width=500)
        self.rightframe.grid(row=0, column=0, padx=(220, 100), pady=40, sticky='nsew')
        self.rightframe.pack_propagate(False)
        self.righttop=ctk.CTkTextbox(self.rightframe, fg_color="#b795ff", height=325, width=450, wrap="word", font=ctk.CTkFont("", 14), text_color="#fafafa", border_color="#4f406e", border_width=3)
        self.righttop.pack(side="top", anchor='center', pady=(20, 2.5))
        def topdata():
            yestdate=dt.datetime.today().date()-dt.timedelta(1)
            try:    
                command="SELECT * FROM journalentries WHERE date=%s"
                Application.c.execute(command, (yestdate,))
                yestrecord=Application.c.fetchall()
                yestentry="".join(yestrecord[0][1])
                self.righttop.delete(0.0, 'end')
                self.righttop.insert(0.0, yestentry)
                self.righttop.configure(state='disable')
            except Exception:
                errortext="NO JOURNALS WERE WRITTEN YESTERDAY"
                self.righttop.delete(0.0, 'end')
                self.righttop.insert(0.0, errortext)
        topdata()
        self.rightbottom=ctk.CTkTextbox(self.rightframe, fg_color="#b795ff", height=325, width=450, wrap='word', font=ctk.CTkFont("", 14), text_color="#fafafa", border_color='#4f406e', border_width=3)
        self.rightbottom.pack(side="bottom", anchor='center', pady=(2.5, 20))
        def bottomdata():
            dbefyestdate=dt.datetime.today().date()-dt.timedelta(2)
            try:    
                command="SELECT * FROM journalentries WHERE date=%s"
                Application.c.execute(command, (dbefyestdate,))
                yestrecord=Application.c.fetchall()
                yestentry="".join(yestrecord[0][1])
                self.rightbottom.delete(0.0, 'end')
                self.rightbottom.insert(0.0, yestentry)
                self.rightbottom.configure(state='disable')
            except Exception:
                errortext="NO JOURNALS WERE WRITTEN DAY BEFORE YESTERDAY"
                self.rightbottom.delete(0.0, 'end')
                self.rightbottom.insert(0.0, errortext)
        bottomdata()
        self.leftframeunder=ctk.CTkFrame(self, fg_color="#7860ac", height=700, width=500)
        self.leftframeunder.grid(row=0, column=1, padx=(0, 100), pady=40, sticky='nsew')
        self.leftframeover=ctk.CTkFrame(self.leftframeunder, fg_color="#b795ff", height=650, width=450, border_color='#4f406e', border_width=3)
        self.leftframeover.pack(anchor='center', padx=25, pady=25)
        self.leftframeover.pack_propagate(False)
        self.today=ctk.CTkLabel(self.leftframeover, text=str(dt.datetime.today().strftime("%A, %d-%b-%Y")), text_color="#fafafa", font=ctk.CTkFont("", 30))
        self.today.pack(pady=20, side='top', anchor='center')
        self.profimg=ctk.CTkLabel(self.leftframeover, text="", image=ctk.CTkImage(Image.open("F:/Code Prerequisites/icon/H.jpg"), size=(240, 240)),)
        self.profimg.pack(pady=20, anchor='center')
        self.helloname=ctk.CTkLabel(self.leftframeover, text="Welcome Back Harsha", text_color="#fafafa", font=ctk.CTkFont("", 30))
        self.helloname.pack(pady=20, anchor='center')

class DropFrame(ctk.CTkFrame):
    def __init__(self, master):
        ctk.CTkFrame.__init__(self, master, fg_color=None, height=500, width=110) #remove fill, top side west anchor
        self.master=master
        self.pack_propagate(False)
        self.homebutton=ctk.CTkButton(self, image=ctk.CTkImage(dark_image=(Image.open("F:/source/get help son/homebtnimg.png"))), text="", command=lambda: HomeFrame(self.master).pack(side='right', fill='both', expand=1), hover=True, fg_color='transparent', width=100, height=75, hover_color="#b795ff")
        self.homebutton.pack()
        self.journalbutton=ctk.CTkButton(self, image=ctk.CTkImage(dark_image=(Image.open("F:/source/get help son/journalbtnimg.png"))), text="", command=lambda: JournalFrame(self.master).pack(side="right", fill="both", expand=1), fg_color='transparent', width=100, height=75, hover_color="#b795ff")
        self.journalbutton.pack()
        self.searchbutton=ctk.CTkButton(self, image=ctk.CTkImage(dark_image=(Image.open("F:/source/get help son/searchbtnimg.png"))), text="", command=lambda: SearchFrame(self.master).pack(side="right", fill="both", expand=1), fg_color='transparent', width=100, height=75, hover_color="#b795ff")
        self.searchbutton.pack()
        self.todolistbutton=ctk.CTkButton(self, image=ctk.CTkImage(dark_image=(Image.open("F:/source/get help son/todobtnimg.png"))), text="", command=lambda: ToDoFrame(self.master).pack(side="right", fill="both", expand=1), fg_color='transparent', width=100, height=75, hover_color="#b795ff")
        self.todolistbutton.pack()
        self.quitbutton=ctk.CTkButton(self, text="Quit", command=self.master.quit, hover=True, fg_color='transparent', width=100, height=75, hover_color="#b795ff")
        self.quitbutton.pack()

class SearchFrame(ctk.CTkFrame):
    def __init__(self, master):
        killallframes()
        self.master=master
        self.menu=MenuFrame(self.master)
        self.menu.pack(side='top', fill='x')
        ctk.CTkFrame.__init__(self, master, fg_color='#3d113d', corner_radius=0)
        self.searchbar=ctk.CTkEntry(self, fg_color='#1f1f1f', corner_radius=30, width=500, font=(ctk.CTkFont('Sans-serif', 30, slant='italic')), justify='center', text_color="#b795ff", border_width=1)
        self.searchbar.pack(pady=(100, 30))
        self.searchbar.bind("<Button-1>", lambda event: self.searchbar.delete(0, 'end'))
        def searchdata(event):
            self.searchresult.configure(state="normal")
            self.searchresult.delete(0.0, 'end')
            try:
                data=wp.wikipedia.summary(self.searchbar.get(), sentences=5)
                self.searchresult.insert(0.0, data)
            except Exception:
                errormsg="Oooppsss!!!, An error occured while searching.\nCheck for any misspelled words in the search bar, or try entering a key word.\nMaybe what you're searching for does not exist."
                self.searchresult.insert(0.0, errormsg)
            self.searchresult.configure(state='disable')
        self.searchbar.bind("<Return>", searchdata)
        self.searchdatabutton=ctk.CTkButton(self, text="Search", fg_color="#1f1f1f", command=lambda:searchdata(""), hover_color="#b795ff", height=40)
        self.searchdatabutton.pack(pady=(10,100))
        self.searchresult=ctk.CTkTextbox(self, border_width=1, fg_color='#1f1f1f', text_color='#a37e49', activate_scrollbars=True, width=800, height=400, font=ctk.CTkFont('Sans-serif', 20), wrap='word', spacing1=30, spacing2=10, spacing3=30, state='disable')
        self.searchresult.pack()

class JournalFrame(ctk.CTkFrame):
    def __init__(self, master):
        killallframes()
        self.master=master
        self.menu=MenuFrame(self.master)
        self.menu.pack(side='top', fill='x')
        self.dateselector=ctk.CTkFrame(self.master, fg_color="#3d113d", height=40, corner_radius=0)
        self.dateselector.pack(side='top', fill='x')
        self.dateselector.grid_propagate(False)
        self.editorbuttons=ctk.CTkFrame(self.master, fg_color="#3d113d", height=40, corner_radius=0)
        self.editorbuttons.pack(side='bottom', fill='x')
        self.editorbuttons.grid_propagate(False)
        self.calendar=tkcal.DateEntry(self.dateselector, font=ctk.CTkFont("", 20), locale="eng_UK", showweeknumbers=False, firstweekday="sunday", weekenddays=[1,7])
        self.calendar.grid(row=0, column=0, pady=5, padx=(250, 50))
        ctk.CTkFrame.__init__(self, master, fg_color="#fafafa", corner_radius=0)
        self.editorwindow=ctk.CTkTextbox(self, fg_color="#fafafa", corner_radius=0, text_color="#ae7e49", wrap='word', font=ctk.CTkFont("", 20), undo=True, autoseparators=True, maxundo=-1)
        #self.editorwindow.configure(startlines=2.0)
        self.editorwindow.pack(fill='both', expand=1)
        def adddate():    
            today=self.calendar.get_date()
            displaydate=today.strftime("%d-%b-%Y")
            self.editorwindow.insert(0.0, displaydate+" :\n\n")
            self.editorwindow.focus_set()
        adddate()
        def entrysave():
            caldate=self.calendar.get_date()
            entered=self.editorwindow.get(0.0, 'end-1c') #'end'
            #entered=entered[:len(entered)-1]
            if entered=="" or entered[0:15]!=caldate.strftime("%d-%b-%Y")+" :\n\n":
                self.editorwindow.delete(0.0, 'end')
                adddate()
                self.editorwindow.insert(self.editorwindow.index('insert'), entered)
                entered=self.editorwindow.get(0.0, 'end-1c')
            if entered==caldate.strftime("%d-%b-%Y")+" :\n\n":
                messagebox.showerror("EMPTY JOURNAL", "CANNOT SAVE AN EMPTY JOURNAL.\nFIND SOMETHING TO DO AND THEN TELL US ABOUT IT")
            else:    
                try:
                    command="INSERT INTO journalentries(date, entry) VALUES(%s, %s)"
                    values=(caldate, entered)
                    Application.c.execute(command, values)
                    Application.db.commit()
                except Exception:
                    command="UPDATE journalentries SET entry=%s WHERE date=%s"
                    values=(entered, caldate)
                    Application.c.execute(command, values)
                    Application.db.commit()
        def entryget():
            caldate=self.calendar.get_date()
            try:    
                command="SELECT * FROM journalentries WHERE date=%s"
                Application.c.execute(command, (caldate,))
                record=Application.c.fetchall()
                preventrytext="".join(record[0][1])
                self.editorwindow.delete(0.0, 'end')
                self.editorwindow.insert(0.0, preventrytext)
                self.editorwindow.focus_set()
            except Exception:
                entryclear("")
                messagebox.showerror("JOURNAL NOT FOUND", "WOW! SUCH EMPTY.\nNO JOURNALS EXIST FOR THIS DAY")
        def entryclear(event):
            self.editorwindow.delete(0.0, 'end')
            adddate()
            self.editorwindow.focus_set()
        def cuttext():
            try:
                self.master.clipboard_clear()
                self.master.clipboard_append(self.editorwindow.selection_get())
                self.editorwindow.delete('sel.first', 'sel.last')
            except Exception:
                pass
        def copytext():
            try:
                self.master.clipboard_clear()
                self.master.clipboard_append(self.editorwindow.selection_get())
            except Exception:
                pass
        def pastetext():
            try:
                self.editorwindow.insert(self.editorwindow.index('insert'), self.master.clipboard_get())
            except Exception:
                pass 
        self.undobtn=ctk.CTkButton(self.editorbuttons, text="Undo", width=0, command=self.editorwindow.edit_undo, fg_color="#b795ff", hover_color="#ae7e49")
        self.undobtn.grid(row=0, column=0, padx=(50,10), pady=5)
        self.redobtn=ctk.CTkButton(self.editorbuttons, text="Redo", width=0, command=self.editorwindow.edit_redo, fg_color="#b795ff", hover_color="#ae7e49")
        self.redobtn.grid(row=0, column=1, padx=10, pady=5)
        self.cutbtn=ctk.CTkButton(self.editorbuttons, text="Cut", width=0, command=cuttext, fg_color="#b795ff", hover_color="#ae7e49")
        self.cutbtn.grid(row=0, column=2, padx=10, pady=5)
        self.copybtn=ctk.CTkButton(self.editorbuttons, text="Copy", width=0, command=copytext, fg_color="#b795ff", hover_color="#ae7e49")
        self.copybtn.grid(row=0, column=3, padx=10, pady=5)
        self.pastebtn=ctk.CTkButton(self.editorbuttons, text="Paste", width=0, command=pastetext, fg_color="#b795ff", hover_color="#ae7e49")
        self.pastebtn.grid(row=0, column=4, padx=10, pady=5)
        self.savebtn=ctk.CTkButton(self.dateselector, text="Save", width=0, command=entrysave, fg_color="#b795ff", hover_color="#ae7e49")
        self.savebtn.grid(row=0, column=1, padx=(50,10), pady=5)
        self.getbtn=ctk.CTkButton(self.dateselector, text="Get", width=0, command=entryget, fg_color="#b795ff", hover_color="#ae7e49")
        self.getbtn.grid(row=0, column=2, padx=10, pady=5)
        self.clearbtn=ctk.CTkButton(self.dateselector, text="Clear", width=0, command=lambda:entryclear(""), fg_color="#b795ff", hover_color="#ae7e49")
        self.clearbtn.grid(row=0, column=3, padx=10, pady=5)
        self.calendar.bind("<<DateEntrySelected>>", lambda event:[entryclear(event)])

class ToDoFrame(ctk.CTkScrollableFrame):
    def __init__(self, master):
        killallframes()
        self.master=master
        self.menu=MenuFrame(self.master)
        self.menu.pack(side='top', fill='x')
        self.listframe=ctk.CTkFrame(self.master, fg_color="#3d113d", corner_radius=0, height=80)
        self.listframe.pack(side='bottom', fill='x', expand=1)
        checkitemlst=[]
        ToDoFrame.r_no=tk.IntVar()
        ToDoFrame.r_no.set(-1)
        itementrylst=[]
        okaybtnlst=[]
        checkvarlst=[]
        ctk.CTkScrollableFrame.__init__(self, master, fg_color="#b795ff", corner_radius=0, height=900)
        def checkchecked():
            for n, i in enumerate(checkvarlst):
                checkingentry=itementrylst[n]
                if i.get()==1:
                    checkingentry.configure(font=ctk.CTkFont("", 20, overstrike=True), text_color="#21BF73")
                    checkingentry.configure(state='disable')
                else:
                    checkingentry.configure(font=ctk.CTkFont("", 20, overstrike=False), text_color="#ee4b2b")
                    checkingentry.configure(state='normal')
        def createlistitem():    
            self.checkvar=tk.IntVar()
            self.checkitem=ctk.CTkCheckBox(self, border_color="#000", corner_radius=5, text="", width=0, hover_color="#8601af", fg_color="#3d113d", variable=self.checkvar, command=lambda:checkchecked())
            self.itementry=ctk.CTkEntry(self, justify='center', text_color="#ee4b2b", fg_color="#fff", border_width=0, font=ctk.CTkFont("", 20), width=300)
            self.okaybutton=ctk.CTkButton(self, width=20, text="↵", text_color="#0f0", fg_color='#1f1f1f', font=ctk.CTkFont("", 20), command=lambda:[self.okaybutton.destroy(), self.itementry.configure(state='disable')])
            checkitemlst.append(self.checkitem)
            checkvarlst.append(self.checkvar)
            itementrylst.append(self.itementry)
            okaybtnlst.append(self.okaybutton)
            ToDoFrame.r_no.set(len(checkitemlst)-1)
            self.itementry.focus_set()
            if ToDoFrame.r_no.get()==0:
                self.checkitem.grid(row=ToDoFrame.r_no.get(), column=0, padx=(150,0), pady=(50,5))
                self.itementry.grid(row=ToDoFrame.r_no.get(), column=1, padx=(0,10), pady=(50,5))
                self.okaybutton.grid(row=ToDoFrame.r_no.get(), column=2, pady=(50, 5))
            else:
                self.checkitem.grid(row=ToDoFrame.r_no.get(), column=0, padx=(150,0), pady=5)
                self.itementry.grid(row=ToDoFrame.r_no.get(), column=1, padx=(0,10), pady=5)
                self.okaybutton.grid(row=ToDoFrame.r_no.get(), column=2, pady=5)
        self.additembutton=ctk.CTkButton(self.listframe, text="+", fg_color='#fafafa', text_color="#ae7e49", font=ctk.CTkFont("", 50), width=60, hover_color="#8601af", command=lambda:[[self.okaybutton.destroy(), createlistitem()] if ToDoFrame.r_no.get()>=0 else createlistitem()])
        self.additembutton.grid(row=0, column=0, pady=10, padx=50)
        self.removeitembutton=ctk.CTkButton(self.listframe, text="–", fg_color='#fafafa', text_color="#ae7e49", font=ctk.CTkFont("", 45, 'bold'), width=60, height=60, hover_color="#8601af", command=lambda:[[okaybtnlst[ToDoFrame.r_no.get()].destroy(), okaybtnlst.pop(ToDoFrame.r_no.get()), itementrylst[ToDoFrame.r_no.get()].destroy(), itementrylst.pop(ToDoFrame.r_no.get()), checkitemlst[ToDoFrame.r_no.get()].destroy(), checkitemlst.pop(ToDoFrame.r_no.get()), checkvarlst.pop(ToDoFrame.r_no.get()), ToDoFrame.r_no.set((ToDoFrame.r_no.get()-1))] if ToDoFrame.r_no.get()>=0 else lambda:""])
        self.removeitembutton.grid(row=0, column=1, pady=10, padx=50)
        self.removeallbutton=ctk.CTkButton(self.listframe, text="⨉", fg_color='#fafafa', text_color="#ae7e49", font=ctk.CTkFont("", 40), width=60, height=60, hover_color="#8601af", command=lambda:[[[okaybtnlst[i].destroy(), okaybtnlst.pop(i), itementrylst[i].destroy(), itementrylst.pop(i), checkitemlst[i].destroy(), checkitemlst.pop(i), checkvarlst.pop(i), ToDoFrame.r_no.set((ToDoFrame.r_no.get()-1))] for i in range(len(checkitemlst)-1, -1, -1)]if ToDoFrame.r_no.get()>=0 else lambda:""])
        self.removeallbutton.grid(row=0, column=2, pady=10, padx=50)

def killallframes():
    for i in root.winfo_children():
        i.destroy()

if __name__=='__main__':
    root=ctk.CTk()
    app=Application(root)
    root.mainloop()
```
